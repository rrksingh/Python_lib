# Bar Graphs
- plt.bar()
- plot a bar graph between heights and labels you provides.

```python
x = np.arange(4)
car1 = [40, 17, 28, 12]
car2 = [34, 21, 24, 15]
```

```python
plt.bar(x,car1,label='car1',tick_label=[2017,2018,2019,2020],color='red')
plt.bar(x,car2,label='car2',color='yellow')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.show()
```

# In this loop:

- i represents the x-coordinate of the bar.
- v represents the height of the bar.
- v + 0.1 sets the y-coordinate for the annotation slightly above the bar.
- str(v) is the text displayed as the value on top of each bar.
- ha='center' horizontally aligns the text to the center of the bar.

```python
plt.bar(x,car1,width=0.3,label='car1',tick_label=[2017,2018,2019,2020])
plt.bar(x+0.3,car2,width=0.3,label='car2')
for i, v in enumerate(car1):
    plt.text(i, v, str(v), ha='center')
for i, v in enumerate(car2):
    plt.text(i+0.3, v + 0.1, str(v), ha='center')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

a = np.linspace(-5, 5, 100)  # Sample data for the x-axis
b = np.linspace(-5, 5, 100)  # Sample data for the y-axis
a, b = np.meshgrid(a, b)  # Create a meshgrid for a 2D surface

Z = np.array([[5]])  # Example data for the z-axis

ax.plot_surface(a, b, Z)

plt.show()
```
