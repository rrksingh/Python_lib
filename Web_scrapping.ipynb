# Beautiful Soup
Beautiful Soup is a popular Python library used for scraping information from web pages. It is renowned for its ease of use and flexibility, allowing developers to parse HTML and XML documents effortlessly.

When scraping a web page, Beautiful Soup can be used to extract specific data elements, such as text, hyperlinks, and other HTML attributes. It enables developers to navigate and search the parse tree created from the web page's source code, making it easier to locate and extract relevant information.

# PROJECT1: Scrapping countries related data from wikipedia site

```python
import bs4
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_and_their_capitals_in_native_languages"
```

```python
res = requests.get(url)
```

```python
if res.status_code==200:
    print("Connection established")
else:
    print("Connection fail")
```

```python
text = res.content
text
```

```python
soup = bs4.BeautifulSoup(text)
```

```python
type(soup)
```

```python
soup.prettify
```

```python
soup.h2.text
```

```python
soup.title.text
```

```python
soup.table
```

```python
lst=soup.find_all('h2')
```

```python
for i in lst:
    print(i.text)
```

```python
tables= soup.find_all('table')
```

```python
len(tables)
```

```python
tables= soup.find_all('table',attrs={'class':'wikitable'})
```

```python
len(tables)
```

# getting information about one country

```python
b_table=tables[1].find_all('tr')[1:]
```

```python
len(b_table)
```

```python
all_tds=b_table[0].find_all('td')
```

```python
country = all_tds[0].text
capital = all_tds[1].text
language = all_tds[-1].text
```

```python
print(country, capital, language)
```

```python
all_tds[0].find('img').attrs['src']
```

```python
df =  pd.DataFrame({},columns=['country', 'capital', 'language'])
```

```python
df
```

```python
for table in tables:
    all_rows = table.find_all('tr')[1:]
    for row in all_rows:
        all_tds= row.find_all('td')
        country = all_tds[0].text
        capital = all_tds[1].text
        language = all_tds[-1].text
        e= {'country':country,'capital':capital,'language':language}
        df=pd.concat([df,pd.DataFrame([e])], ignore_index=True)
```

```python
df
```

```python
df['language']=df['language'].apply(lambda x :x[:-1])
```

```python
df
```

```python
df.to_csv('country.csv',index=False)
```

# PROJECT2:Scraping Inspirational Quotes from site :  -  www.passiton.com/inspirational-quotes

```python
import requests
import bs4
```

```python
# url = "https://www.passiton.com/inspirational-quotes"
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
res = requests.get(url)
if res.status_code==200:
    print("Connection established")
else:
    print("Connection fail")
```

```python
data = res.content
soup = bs4.BeautifulSoup(data)
soup
```

```python
box = soup.find("div", attrs={"id":"all_quotes"})
box
```

```python
imgs =box.find_all("img")
```

```python
len(imgs)
```

```python
imgs
```

```python
res = requests.get(img_url)
```

```python
img_content = res.content
```

<h2>scraping all the images</h2>

```python
for idx,i in enumerate(imgs,1):
    img_url = i.attrs['src']
    
    res = requests.get(img_url)
    img_content = res.content
    
    with open(f"images\quotes_{idx}.jpg", 'wb') as f:
        f.write(img_content)
```

<h2>scraping all the images from first 3 pages</h2>

```python
j=1
for k in range(1,4):
    if k==1:
        url = "https://www.passiton.com/inspirational-quotes"
    else:
        url = f"https://www.passiton.com/inspirational-quotes?page={k}"
    r = requests.get(url)
    soup = bs4.BeautifulSoup(r.content)
    box = soup.find("div", attrs={"id":"all_quotes"})
    imgs =box.find_all("img")
    for i in imgs:
        img_url = i.attrs['src']

        res = requests.get(img_url)
        img_content = res.content

        with open(f"images\quotes_{j}.jpg", 'wb') as f:
            f.write(img_content)
        j+=1
```
